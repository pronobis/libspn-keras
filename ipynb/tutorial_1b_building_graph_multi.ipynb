# Tutorial 1b: Building the SPN Graph Using Multi-Op Nodes

```python
import libspn as spn
import tensorflow as tf
```

### Build

```python
indicators_x = spn.IndicatorLeaf(num_vars=2, num_vals=2, name="indicators_x")
sums_1 = spn.ParSums((indicators_x, [0,1]), num_sums=2, name="sums_1")
sums_2 = spn.ParSums((indicators_x, [2,3]), num_sums=2, name="sums_2")
prods_1 = spn.PermProducts(sums_1, sums_2, name="prods_1")
root = spn.Sum(prods_1, name="root")
latent_indicator_y = root.generate_latent_indicators(name="latent_indicator_y") # Can be added manually
spn.generate_weights(root, initializer=tf.initializers.random_uniform()) # Can be added manually
```

### Inspect

```python
print(root.get_num_nodes())
print(root.get_scope())
print(root.is_valid())
```

### Visualize the SPN Graph

```python
spn.display_spn_graph(root)
```
